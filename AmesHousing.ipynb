##**Correlation Heatmap of Ames Dataset**

```python
plt.figure(figsize=(30, 30))
#The values to be displayed in color map from -1 to 1. annot is used to give the correlation values
heatmap = sns.heatmap(csvdata.corr(), vmin=-1, vmax=1, annot=True, cmap='BrBG')
# Setting the heat map title
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':20}, pad=15);
# dpi - sets the resolution of the saved image in dots/inches
# bbox_inches - when set to 'tight' - does not allow the labels to be cropped
plt.savefig('heatmap.png', dpi=300, bbox_inches='tight')
```

The values in the boxes indicate the correlation between each variable in the axis. Correlation values ranges from -1 to 1. Those values which are closer to 0 indicate no linear trend, while those which are closer to 1 indicate strong correlation.

When the correlation value approaches 1, we can say that both values in each axis are increasing. In the same way, if the value is close to -1, the value in one axis is increasing and the value in the other axis is decreasing.

# **Data Cleaning**

## Dropping Columns

```python
#.drop(columns=['']) function is used to drop the columns.
# axis=1 represents the names of othe axis with the following column names to drop
csvdata = csvdata.drop(columns=['Order','Alley','Fireplace Qu','Pool QC','Fence','Misc Feature'],axis=1)
csvdata
```

In the above step, I have dropped the columns which have more number of missing values i.e. greater than 50% of the data. So, it will be helpfull to perform data exploration.

##Data imputation

```python
csv_data1 = csvdata.fillna(csvdata.mode().iloc[0])
csv_data1
```

.fillna() function is used to replace all the missing values.
 I have used mode because it will give the highest categorial value in place of missing data. The mean or median will only replace the numerical value of the dataset.

Checking is there are any null values after replacing the null values with the mode

```python
#.isnull().sum().to_string() will return the sum of the total number of missing values
print(csv_data1.isnull().sum().to_string())
```

Created a heat map to check if there are any null values and I could find that there are no null values listed in the heatmap

```python
# "yticklabels=False" represents do not plot the column names if they are false on y-axis
# "cbar=False" to draw a color bar if condition is false
# "cmap='coolwarm'" represents the colour of the map
sns.heatmap(csv_data1.isnull(),yticklabels=False,cbar=False,cmap='coolwarm')
```

#**Exploratory Data Analysis and Visualization**

```python
# sns.distplot is used to shows a histogram with a reference line
sns.distplot(csv_data1['SalePrice'])
```

This above plot is a right-skewed distribution, also known as asymmetrical distribution, since the SalePrice is not normal. There are only a few houses with a higher SalePrice, and the majority of them are lower priced homes.

```python
import numpy as np
# np.log is used to calculate the logarithmic value
target=np.log(csv_data1['SalePrice'])
sns.distplot(target)
```

Most of the algorithms will work effectively, when the data is normally distributed. I have used a target variable by using log scale. Now, the SalePrice variable is more normally distributed than it was earlier.

```python
numerical_data = csv_data1.select_dtypes(include = [np.number])
categorial_data = csv_data1.select_dtypes(exclude = [np.number])

print('In the dataset, there are {0} numerical columns and {1} categorial columns'.
      format(numerical_data.shape[1], categorial_data.shape[1]))
```

In the above cell, To find the numerical and categorial columns in the dataset. I have printed the number of numerical and categorial columns respectively.

Now, lets calculate the correlation of the numerical data

```python
corr = numerical_data.corr()
#sns.heatmap(corr) function is used to plot correlation map
sns.heatmap(corr)
```

The white colour diagonal line in the figure represents the data which has correlation value of 1. The lighter colour represtents the data is more correlated i.e these values have values nearer to 1.

This principle is called Multi Collinearity. It is defined as "It occurs when there are good correlations between two or more variables". This means an independent variable is strongly correlated with other independent variables

### Numerical data distribution

```python
a = pd.melt(csv_data1,value_vars=sorted(numerical_data))
b = sns.FacetGrid(a, col='variable', col_wrap=4, sharex=False, sharey=False)
b = b.map(sns.histplot, 'value')
```

To plot the numerical features and their distribution, I used melt method.

The data distribution in the first figure is positively skewed and has a lower value for the 1st floor SF. It is the same for the Garage area, the Great Room area, and the Basement square footage.

Bars appear in some figures like overall Quall, TotRMSAbvGrd, MoSold, and YrSold to indicate they are not continuous values even though they have numerical values.

###Analyzing the categorical features

```python
#Count plots for means of categorical features
a = pd.melt(csv_data1,id_vars=['SalePrice'],value_vars=sorted(categorial_data))
b = sns.FacetGrid(a, col='variable', col_wrap=3, sharex=False, sharey=False, size=5)
b = b.map(sns.boxplot, 'value', 'SalePrice')
[plt.setp(ax.get_xticklabels(), rotation=90) for ax in b.axes.flat]
b.fig.tight_layout()
plt.show()
```

We can find the difference between the means for each plot from the box plots. The purpose of this plot is to determine which category of dataset has significant mean differences.

It has been noted that the mean of the data in Condition 2 Boxplot differ somewhat in that PosA records the highest with a count of over 400000 and RRNn records the lowest with a count of 100000.

There are also differences between the means in other plots such as neighbourhood, sale type, kitchen quality, etc.

```python
predictors = csv_data1.corr()[['SalePrice']].sort_values('SalePrice', ascending=False)
corr_df = pd.DataFrame(predictors.iloc[1:10])

plt.figure(figsize=(10, 8))
sns.barplot( corr_df['SalePrice'], corr_df.index, palette='bright')
plt.title('Top 10 Predictors of House Price', fontsize=18)
plt.xlabel('Correlation w/SalePrice', fontsize=14)
plt.ylabel('Variable', fontsize=14)
plt.show()
```

To find out which factors of dataset are most helpful to predict house prices, I plotted a bar chart. Overall Qual has the highest dependency rate on the prediction of house price.
